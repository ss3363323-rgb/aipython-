```python
cafe_menu = {
    "아메리카노": 4500,
    "카페라떼": 5000,
    "바닐라라떼": 5500,
    "카푸치노": 5000,
    "치즈케이크": 6900
}

order_lines = []   # (메뉴이름, 수량, 단가, 금액)
total = 0

# 입력) 여러 번 주문 (엔터만 치면 종료)
while True:
    menu = input("메뉴를 선택해주세요(종료: 엔터): ").strip()
    if menu == "":
        break

    qty_str = input("수량을 입력해주세요: ").strip()

    # 처리) 유효성 검사
    if menu not in cafe_menu:
        print("해당 메뉴는 없습니다.\n")
        continue
    if not qty_str.isdigit() or int(qty_str) <= 0:
        print("수량은 1 이상의 숫자로 입력해주세요.\n")
        continue

    qty = int(qty_str)
    unit_price = cafe_menu[menu]
    price = unit_price * qty

    order_lines.append((menu, qty, unit_price, price))
    total += price
    print("장바구니에 담았습니다.\n")

# 출력)
if not order_lines:
    print("주문이 없습니다.")
else:
    for menu, qty, unit_price, price in order_lines:
        print(f"{menu} * {qty}")
    print("---------")
    print(f"총금액 {total}원")
```

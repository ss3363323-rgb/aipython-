```python
num1 = 12                # 첫 번째 숫자(피연산자)를 저장할 변수가 필요해서 만듦
num2 = 5                 # 두 번째 숫자(피연산자)도 저장해야 하므로 변수를 하나 더 만듦

print("12 + 5 =", num1 + num2)   # + 연산 결과를 사람이 읽기 쉽게 “식 + 결과” 형태로 출력하려고 필요
print("12 - 5 =", num1 - num2)   # - 연산 결과를 같은 방식으로 확인하려고 필요
print("12 * 5 =", num1 * num2)   # * 연산 결과를 출력하려고 필요
print("12 / 5 =", num1 / num2)   # / 연산 결과를 출력하려고 필요(나눗셈은 소수로 나올 수 있음)
```

```python
count=0
while user != answer:
  if user < answer:
    print("더 큰 값을 입력하시오")
  else:
    print("더 작은 값을 입력하시오")
  count+=1
  user=int(input("추측값을 입력하시오"))
print("정답입니다", count+1, "번 째에 성공하셨습니다")
```

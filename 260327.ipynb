```python
import random

dice1=random.randint(1,6)
dice2=random.randint(1,6)

print(f"주사위1: {dice1}")
print(f"주사위2: {dice2}")

if dice1==dice2:
  print("비겼습니다")
elif dice1>dice2:
  print("주사위1이 이겼습니다")
else:
  print("주사위2가 이겼습니다")
```

```python
#가위바위보
import random

man1=random.randint(1,3)
man2=random.randint(1,3)

print(f"사람 1은 {man1}입니다")
print(f"사람 2은 {man2}입니다")

if man1==man2:
  print("비겼습니다")
elif (man1==1 and man2==3)or(man1==2 and man2==1)or(man1==3 and man2==2):
  print("사람 1이 이겼습니다")
else:
  print("사람 2가 이겼습니다")

```

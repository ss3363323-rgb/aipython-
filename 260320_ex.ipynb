```python
# 사용자에게 정수를 입력받아 짝수/홀수 판별 후 출력

raw = input("숫자를 입력하세요: ").strip()

try:
    n = int(raw)
except ValueError:
    print("정수만 입력해 주세요. 예: 10")
else:
    if n > 0:
        print(f"{n}은(는) 양수입니다.")
    else:
        print(f"{n}은(는) 음수입니다.")
```

```python
COF_price=2000
LAT_price=3000
GRE_price=2500
JUI_price=2800

COF_lot = int(input())
LAT_lot = int(input())
GRE_lot = int(input())
JUI_lot = int(input())

A = COF_price * COF_lot
B = LAT_price * LAT_lot
C = GRE_price * GRE_lot
D = JUI_price * JUI_lot

print('커피주문수: ', COF_lot)
print('라떼주문수: ', LAT_lot)
print('녹차주문수: ', GRE_lot)
print('쥬스주문수: ', JUI_lot)
print('총주문금액: ', A+B+C+D, '원')
```
